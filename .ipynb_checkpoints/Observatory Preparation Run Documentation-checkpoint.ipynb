# Observing Run Preparation


## PROGRAM :
>Observation Run Preparation
## AUTHOR :
>Aaron Philip, Saket Kulkarni, Vinit Hardikar, Gaurang Potdar, Krishnaraj Thadesar, Krishna Kapoor
## PURPOSE :
>The purpose of Observing Run Preparation is to maximise the use of the time alloted to us for observing these celestial bodies in the most effective way.
## OUTPUT :
>A spreadsheet with the details of the Target, RA, DEC, Rise Time, Transit,Set Time  and Observability of a celestial body

## Main Method :
> **MAIN FUNCTION** Main method is the function from where a program effectively starts. It is essentially the command center for a program from where the program takes its due course of action. <br>
>  **INPUT** The input is The input to the main function is acquired by '.json' files which comprise of the follwoing input functions - <br>
i) `constraints.json` - Includes upper value of Airmass, Altitude and moon seperation and lower acceptable value of minimum altitude. It also states the start and end time of the observation period. <br>
ii)`date_and_time.json`- Houses information of current date and time. If customised time is required it can also be input. <br>
iii)`observatory.json` - Houses information about name, description, latitude, longitude, elevation and timezone of the observatory. <br>
iv) `target.json`- Stores names of targets in an array format of string type. <br>
Another program `parse_input.py` converts the .json files into pyhton acceptable input and calculates and fetches various values. <br>
> **OUTPUT** The output is generated as a '.csv' format in two sheets. The outputs are -    <br>
i)`observer_info.csv` - Showcases the date and time of nearest Sunset Time, Evening Twilight, Midnight Time, Morning Twilight Time and Sunrise Time in a table format.   <br>
ii)`targets_info.csv` - Showcases table consisting of with the following information -  <br> 
TARGET - Name of observable target,RA- Right Ascension,DEC - Declination ,RISE TIME - Rise time of target,SET TIME - sunset time of target ,TRANSIT (Hours) - Transit time of target in hours ,OBSERVABLE DURING TRANSIT - Whether target is observable during target or not.  <br>

>**FUNCTION:** `moon()` <br>
 **ARGUMENTS:** none <br>
 **RETURNS:** Location of Moon at the instant. <br>
 **NOTES:** Moon contributes to light pollution in night sky and therefore its position needs to be taken into account for accurate readings of other celestial objects. <br>

 **ARGUMENTS:** none <br>

>**FUNCTION:** `airmass()` <br>

`time_of_observation`- Time details of night sky  <br>         `target` - Position of celestial object to be observed. <br>
**RETURNS:** Airmass index for the night. <br>
 **NOTES:** Airmass contributes to refraction of incident light and therfore result in inaccurate readings. To counter such errors airmass needs to be taken in consideration. The following is also printed out. As a general rule of thumb, try observing targets when airmass > 2. <br>

** In main **
 
>**FUNCTION:** `sunset()` <br>
 **ARGUMENTS:** `observatory=observatory_setup()` - Object of observatory_setup() is passed. <br>
 **RETURNS:** Exact time of sunset at the location. <br>
 **NOTES:** Determines the exact sunset time, in local time, at the location as celestial bodies can only be observed in the nightsky. <br>

>**FUNCTION:** `evening_twilight()` <br>
 **ARGUMENTS:** `observatory=observatory_setup()` Object of observatory_setup() is passed. <br>
 **RETURNS:** Exact time of evening twilight. <br>
 **NOTES:** Evening twilight is the faint sunlight seen even after the Sun is below the horizon. The time interval of it is fetched by the function. <br>

>**FUNCTION:** `midnight()` <br> 
 **ARGUMENTS:** `observatory=observatory_setup()` Object of observatory_setup() is passed. <br> 
 **RETURNS:** The Zeroth hour (midnight) is returned. <br> 
 **NOTES:** Midnight (00:00) is translated in local time and retunred in this function. <br>

>**FUNCTION:** `morn_twilight()`
 **ARGUMENTS:** `observatory=observatory_setup()` Object of observatory_setup() is passed. <br> 
 **RETURNS:** Exact time of evening twilight. <br>
 **NOTES:** Morning twilight is the faint sunlight seen even before the Sun rises the horizon. The time interval of it is fetched by the function. <br>
 
>**FUNCTION:** `sunrise()`  <br> 
 **ARGUMENTS:** `observatory=observatory_setup()` Object of observatory_setup() is passed. <br> 
 **RETURNS:** Exact time of sunrise at the location. <br> 
 **NOTES:** Determines the exact sunrise time, in local time, at the location. The telescopes should be protected from sunlight as it may damage the light sensitive lenses. <br>  

>**FUNCTION:** `local_siderial_time()` <br>
 **ARGUMENTS:** `observatory=observatory_setup()` Object of observatory_setup() is passed. <br>
 **RETURNS:** The sidereal timing of the observatory. <br>
 **NOTES:** As the Earth rotates on its axis, the celestial sphere appears to revolve around the Earth (on true axis), making one complete revolution in one sidereal day. This helps in relational observation with respect to other celestial object and is determined by this function. <br>
